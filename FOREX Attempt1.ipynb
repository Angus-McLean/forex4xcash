## Imports

```python
import pandas, numpy as np
import matplotlib.pyplot as plt
```

```python
# colTypes = {'VALUE': 'float64'}
rawCan = pandas.read_csv(baseDir + 'DEXCAUS.csv')
rawCan.describe()
```

```python
rawCan.info()
```

```python
rawCan = rawCan.replace('.', np.nan).fillna(method='ffill')
rawCan.VALUE = rawCan.VALUE.astype(float)
```

```python
plt.plot(rawCan['VALUE'][::10])
```

## Moving Averages

```python
windows = [20,50,200]
for i in windows:
    rawCan[str(i)+'_AVG'] = rawCan.VALUE.rolling(i).mean()
```

```python
rawCan
```

```python
plt.plot(
    rawCan['DATE'][-1000::10],
    rawCan['VALUE'][-1000::10],
    rawCan['DATE'][-1000::10],
    rawCan['20_AVG'][-1000::10],
    rawCan['DATE'][-1000::10],
    rawCan['50_AVG'][-1000::10],
    rawCan['DATE'][-1000::10],
    rawCan['200_AVG'][-1000::10]
)
```

## Identifying Trends

```python
rawCan['UpTrend'] = rawCan['20_AVG'] > rawCan['50_AVG']
```

```python
upIndexs = []
downIndexs = []
for ind, row in rawCan.iterrows():
    if row['UpTrend'] and not rawCan.iloc[ind-1]['UpTrend']:
        upIndexs.append(ind)
    elif not row['UpTrend'] and rawCan.iloc[ind-1]['UpTrend']:
        downIndexs.append(ind)
```

```python
print len(upIndexs)
print len(downIndexs)
```

```python
zip(upIndexs, downIndexs)
```

```python
zip(downIndexs[:-1], upIndexs[1:])
```

## Place Trades

```python
trades = map(lambda (upInd, downInd) : rawCan.iloc[upInd]['VALUE']-rawCan.iloc[downInd]['VALUE'], np.array(zip(upIndexs, downIndexs)))
```

```python
trades = map(lambda (upInd, downInd) : rawCan.iloc[upInd]['VALUE']-rawCan.iloc[downInd]['VALUE'], np.array(zip(downIndexs[:-1], upIndexs[1:])))
```

```python
print "Percentage Increase : "
np.sum(trades) / np.average(rawCan.VALUE)
```
